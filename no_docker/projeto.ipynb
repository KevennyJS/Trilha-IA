```python
import random

import numpy
from default import *
from moinhos import moinhos
from verificarJogo import *
from verificarMovimento import *
from verificarCaminho import *
from verificarMoinho import *


def proximoJogador(jogador=None):
    if jogador == 0:
        return 1
    else:
        return 0


def main():
    changePlayer = False
    movimentarPara = 0
    jogador = 0
    listaMoinhosContidos = []
    listMoinhosPossiveis =[]
    pecasJogador1 = 8
    pecasJogador2 = 8
    contador_movimentor_endgame = 0
    bloquearVerificacao = False
    while True:
        changePlayer = False
        print(f"Jogador {jogador}\n\n")
        print(f"{tabuleiro} \n\n")

        if pecasJogador1 > 0 and pecasJogador2 > 0:  # se ainda houver peças para serem jogadas pelos usuarios
            if jogador == 0:
                pecaEscolhida = int(input("Digite o campo escolhido: "))
                if tabuleiro[pecaEscolhida] is None and pecaEscolhida < 24:
                    if jogador == 0:
                        tabuleiro[pecaEscolhida] = jogador
                        pecasJogador1 -= 1
                        changePlayer = True
                else:
                    bloquearVerificacao = True
                    print("O campo escolhido já possui uma peça, por favor, escolha outro campo.")
            else:
                while True:
                    pecaEscolhida = iaEscolha = random.randint(0, 23)
                    if (tabuleiro[iaEscolha] is None):
                        tabuleiro[iaEscolha] = jogador
                        pecasJogador2 -= 1
                        changePlayer = True
                        break


        else:
            if(jogador == 1):
                for i in range(24):
                    if tabuleiro[i] == 1:
                        #se encontrar peças do bot no tabuleiro
                        for element in moinhos:  # procurar combinações que possuam a nova posição da peça
                            listaMoinhoscontidos = []
                            if element.__contains__(i):
                                if not listaMoinhosContidos.__contains__(element):
                                    listaMoinhosContidos.append(element)
                        for moinhoFiltrado in listaMoinhosContidos:
                            if len(moinhoFiltrado)>0:
                                if ((tabuleiro[moinhoFiltrado.__getitem__(0)] == None or tabuleiro[moinhoFiltrado.__getitem__(0)] == 1) and 
                                    (tabuleiro[moinhoFiltrado.__getitem__(1)] == None or tabuleiro[moinhoFiltrado.__getitem__(1)] == 1) and 
                                    (tabuleiro[moinhoFiltrado.__getitem__(2)] == None or tabuleiro[moinhoFiltrado.__getitem__(2)] == 1)):
                                        if not (tabuleiro[moinhoFiltrado.__getitem__(0)] == 1 
                                           and tabuleiro[moinhoFiltrado.__getitem__(1)] == 1 
                                           and tabuleiro[moinhoFiltrado.__getitem__(2)] == 1):
                                            listMoinhosPossiveis.append(moinhoFiltrado)   
                quantidade_pecas_contidas = []
                for atual in listMoinhosPossiveis:
                    contador = 0
                    for index in range(3):
                        if tabuleiro[atual.__getitem__(index)] == 1: 
                            contador = +1
                #max()
                index_maior = 0        
                for x in range(len(quantidade_pecas_contidas)):
                    if quantidade_pecas_contidas[x] > index_maior:
                        index_maior = quantidade_pecas_contidas[x].numerator
                
                moinho_escolhido = listMoinhosPossiveis[index_maior]
                pecas_nao_none = []
                #escolhendo um campo do moinho escolhido, para ser o objetivo da busca
                if not tabuleiro[moinho_escolhido.__getitem__(0)] == 1:
                    peca_objetivo = moinho_escolhido.__getitem__(0)
                    pecas_nao_none.append(moinho_escolhido.__getitem__(0))
                elif not tabuleiro[moinho_escolhido.__getitem__(1)] == 1:
                    peca_objetivo = moinho_escolhido.__getitem__(1)
                    pecas_nao_none.append(moinho_escolhido.__getitem__(1))
                elif not tabuleiro[moinho_escolhido.__getitem__(2)] == 1:
                    peca_objetivo = moinho_escolhido.__getitem__(2)
                    pecas_nao_none.append(moinho_escolhido.__getitem__(2))
                
                caminhos_todos = []
                # buscando caminho de todas as peças disponiveis para ir até o objetivo
                print("Iniciando buscas:\n")
                for i in range(24):
                    #verifica se a peça que ele esta lendo é do jogador 1 e não é uma peça do moinho ja escolhido
                    if tabuleiro[i] == 1 and not pecas_nao_none.__contains__(i):
                        print(verificarCaminho(f"q{i}",f"q{peca_objetivo}"))
                        #caminhos_todos.append()
                        
                        
                print(f"Todos os caminhos: {caminhos_todos}")
            pecaEscolhida = int(input("Selecione a peça que deseja movimentar: "))
            movimentarPara = int(input("Digite o campo para onde deseja movimentar: "))
            if tabuleiro[pecaEscolhida] == jogador:
                # verifica se o jogador atual tem 3 peças, para poder movimentar para qualquer lugar do mapa e se o campo para onde ele vai está vazio
                if tabuleiro.count(jogador) == 3 and tabuleiro[movimentarPara] is None:
                    if contador_movimentor_endgame < 10 and tabuleiro.count(1) == 3 and tabuleiro.count(0) == 3:
                        tabuleiro[movimentarPara] = jogador  # salva a nova posição da peça
                        tabuleiro[pecaEscolhida] = None  # limpa a posição antiga da peça
                        contador_movimentor_endgame += 1
                        changePlayer = True  # muda o jogador
                        status_do_game = verificarJogo()  # verifica se alguém venceu
                        if status_do_game != 2:
                            if status_do_game == 1:
                                print("Jogador 2 Venceu")
                            else:
                                print("Jogador 1 Venceu")
                            exit()
                    elif contador_movimentor_endgame == 10:
                        print("Fim de jogo! Empate.")
                        exit()
                    else:
                        tabuleiro[pecaEscolhida] = None
                        tabuleiro[movimentarPara] = jogador
                        status_do_game = verificarJogo()  # verifica se alguém venceu
                        if status_do_game != 2:
                            if status_do_game == 1:
                                print("Jogador 2 Venceu")
                            else:
                                print("Jogador 1 Venceu")
                            exit()
                elif verificarmovimento(pecaEscolhida, movimentarPara) and tabuleiro[movimentarPara] is None:  # movimentação comum
                    tabuleiro[movimentarPara] = tabuleiro[pecaEscolhida]  # salva a nova posição da peça
                    tabuleiro[pecaEscolhida] = None  # limpa a posição antiga da peça
                    changePlayer = True #muda o jogador
                    status_do_game = verificarJogo() #verifica se alguém venceu
                    if status_do_game != 2:
                        if status_do_game == 1:
                            print("Jogador 2 Venceu")
                        else:
                            print("Jogador 1 Venceu")
                        exit()
                else:
                    bloquearVerificacao = True
                    print("Você não pode mover para uma casa onde já possue peças")
            else:
                bloquearVerificacao = True
                print("Essa peça não é sua, por favor, escolha outra")

        if bloquearVerificacao is False:
            listaMoinhosContidos = []  # limpando a lsita de moinhos que contem a peça selecionada
            for element in moinhos:  # procurar combinações que possuam a nova posição da peça
                if pecasJogador1 > 0 and pecasJogador2 > 0:
                    if element.__contains__(pecaEscolhida):
                        listaMoinhosContidos.append(element)
                else:
                    if element.__contains__(movimentarPara):
                        listaMoinhosContidos.append(element)
            print(listaMoinhosContidos)
            for filtered in listaMoinhosContidos:  # verificando se entre as separadas, alguma da "match" de moinho
                if tabuleiro[filtered[0]] == jogador and tabuleiro[int(filtered[1])] == jogador and tabuleiro[int(filtered[2])] == jogador:
                    print("===================MOINHO====================")
                    if jogador == 0:
                        pecaRemover = int(input("Selecione uma peça do adversario para remover: "))
                        if tabuleiro[pecaRemover] != jogador and tabuleiro[pecaRemover] != None:
                            tabuleiro[pecaRemover] = None
                            break
                    else:
                        while True:
                            aleatorio = random.randint(0,23)
                            if(tabuleiro[aleatorio] == 0):
                                tabuleiro[aleatorio] = None
                                print(f"A IA removeu a peça {aleatorio}")
                                break

            if changePlayer:
                jogador = proximoJogador(jogador)
        bloquearVerificacao = False


if __name__ == "__main__":
    main()
```
